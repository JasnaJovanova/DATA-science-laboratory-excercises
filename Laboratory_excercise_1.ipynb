### Requirements
Import libraries and modules that you are going to use

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
from urllib.parse import urljoin

```

### Send HTTP request to the target





```python
base_url = "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/"
```

check the response status code

```python
response = requests.get(base_url)
response.status_code
```

### Parse the response content with BeautifulSoap

```python
bs = BeautifulSoup(response.text, "html.parser")
```

### Extract data from the BeautifulSoap object using any selectors, attribute identifiers, etc.



*    Product title (selector `'.wd-entities-title'`)

*  
Product regular price (selector `'.woocommerce-Price-amount'`)

*   Product discount price (if available), same selector as regular price

*   URL to the product page

*   Add to cart button URL



```python
# Основен URL
base_url = "https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/"
# URL на страната за производи
page_url = f'{base_url}page/1/'

# Преземање на HTML содржината од страната
website = requests.get(page_url)

if website.status_code == 200:
    bs = BeautifulSoup(website.text, "html.parser")

    # Наоѓање на производите на страницата
    data = bs.find_all("div", {"class": "product-wrapper"})

    products = []

    for item in data:
        # Извлекување на насловот на производот
        title = item.select_one('.wd-entities-title').text.strip()

        # Извлекување на редовната цена
        reg_price = item.select_one('.woocommerce-Price-amount').text.strip()

        # Извлекување на цената со попуст (ако постои)
        prices = item.select('.woocommerce-Price-amount')
        disc_price = prices[1].text.strip() if len(prices) > 1 else None

        # URL на страницата за производот
        product_url = item.select_one('.wd-entities-title a')['href']

        # URL за додавање во кошничка
        add_button_url = item.find("a", {"class": "add_to_cart_button"})["href"]

        # Додавање на податоците за секој производ во листата
        products.append((title, reg_price, disc_price, product_url, urljoin(base_url, add_button_url)))

df=pd.DataFrame(products, columns=["title","regular_price","discount_price","product_url","add_button_url"])
df
```

Repeat the extraction process for each page of products

```python
# Наоѓање на бројот на страници
pages_element = bs.find("ul", {"class": "page-numbers"})
pages_num = len(pages_element.find_all("li")) if pages_element else 1  # Ако не се најде, постави 1

products = []

# Циклус за секоја страна
for page in range(1, pages_num + 1):
    website = requests.get(f'https://clevershop.mk/product-category/mobilni-laptopi-i-tableti/page/{page}/?per_page=36')

    if website.status_code == 200:
        bs = BeautifulSoup(website.text, "html.parser")
        data = bs.find_all("div", {"class": "product-wrapper"})

        for item in data:
            title = item.find("h3", {"class": "wd-entities-title"}).text.strip()
            url = item.find("h3", {"class": "wd-entities-title"}).find("a")["href"]
            butt_url = item.find("a", {"class": "add_to_cart_button"})["href"]

            # Комбинирање на основниот URL со релативниот URL за копчето
            full_butt_url = urljoin(base_url, butt_url)

            prices = item.find("span", {"class": "price"}).find_all("span", {"class": "amount"})
            reg_price = prices[0].find("bdi").text.strip()
            disc_price = None
            if len(prices) > 1:
                disc_price = prices[1].find("bdi").text.strip()

            # Додавање на податоците за секој производ во листата
            products.append((title, reg_price, disc_price, url, full_butt_url))
```

### Create a pandas DataFrame with the scraped products


```python
df = pd.DataFrame(products,columns=["title","regular_price","discount_price","product_url","add_button_url"])
df
```
